```python
# Paso 1: Instalación de dependencias
!pip install tensorflowjs
!pip install tensorflow==2.16.0rc0

# Paso 2: Configuración del entorno
import os
os.environ["TF_USE_LEGACY_KERAS"] = "1"

# Paso 3: Crear carpeta para los archivos generados
!mkdir historial_crediticio

# Paso 4: Importar librerías
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.callbacks import EarlyStopping

# Paso 5: Definir datos de entrada y salida mayores datos mejor precision ya que con pocos no me arrojaba datos correctos
entradas = np.array([
    [3000, 1000, 12],
    [4000, 500, 24],
    [2500, 1500, 6],
    [5000, 0, 36],
    [3500, 200, 48],
    [6000, 700, 12],
    [4500, 1300, 24],
    [10000, 100, 1],
    [5000, 300, 12],
    [1000, 5000, 6],
    [7000, 500, 12],
    [2000, 3000, 24]
], dtype=float)

historico_crediticio = np.array([1, 1, 0, 1, 1, 1, 0, 1, 1, 0, 1, 0], dtype=float)
scaler = MinMaxScaler()
entradas_escaladas = scaler.fit_transform(entradas)
X_train, X_val, y_train, y_val = train_test_split(entradas_escaladas, historico_crediticio, test_size=0.2, random_state=42)
modelo = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=[3]),
    tf.keras.layers.Dense(units=64, activation='relu'),  # Aumentar neuronas
    tf.keras.layers.Dense(units=32, activation='relu'),
    tf.keras.layers.Dense(units=16, activation='relu'),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])

modelo.compile(optimizer=tf.keras.optimizers.Adam(0.001), loss='binary_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

print("Comenzando entrenamiento...")
historial = modelo.fit(X_train, y_train, epochs=1000, validation_data=(X_val, y_val), verbose=False, callbacks=[early_stopping])
print("Modelo entrenado!")

loss, accuracy = modelo.evaluate(X_val, y_val)
print(f'Precisión en el conjunto de validación: {accuracy * 100:.2f}%')

modelo.save('historial_crediticio/historial_crediticio.h5')
!tensorflowjs_converter --input_format=keras historial_crediticio/historial_crediticio.h5 historial_crediticio


entradas_de_prueba = np.array([
    [10000, 100, 1],
    [3000, 1500, 12],
    [2500, 3000, 6],
    [5000, 0, 24],
])

# Escalar las entradas de prueba
entradas_de_prueba_escaladas = scaler.transform(entradas_de_prueba)

# Realizar las predicciones
predicciones = modelo.predict(entradas_de_prueba_escaladas)

# rresultados si es bueno o malo
for entrada, prediccion in zip(entradas_de_prueba, predicciones):
    estado = "Bueno" if prediccion[0] > 0.5 else "Malo"
    print(f"Entrada: {entrada} -> Predicción: Historial Crediticio {estado}")
```
